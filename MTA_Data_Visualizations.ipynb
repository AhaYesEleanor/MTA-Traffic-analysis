```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pickle
%matplotlib inline
```

```python
with open('Benson_full_analysis pickle_v2 .pickle','rb') as read_file:
    weekly_totals = pickle.load(read_file)
```

```python
weekly_totals.sort_values('ENTRIES', ascending=False, inplace=True)
weekly_totals = weekly_totals.reset_index()
weekly_totals.head()
```

```python
top_weekly_totals = weekly_totals[0:10]
```

```python
top_weekly_totals
```

```python
import seaborn as sns
sns.set_color_codes("pastel")
sns.barplot(y=top_weekly_totals['STATION'], x=top_weekly_totals['ENTRIES'], color = "b");
```

```python
top_weekly_totals['cumulative'] = top_weekly_totals['mean_percentage'].cumsum()
top_weekly_totals
```

```python
sns.barplot(y=top_weekly_totals['STATION'], x=top_weekly_totals['cumulative'], color = "b");
plt.xlabel('Cumulative %')
```
