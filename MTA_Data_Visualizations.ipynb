```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pickle
%matplotlib inline
```

```python
with open('final_station_priorities_v2.pickle','rb') as read_file:
    station_priorities = pickle.load(read_file)
```

```python
station_priorities.sort_values('avg_entries', ascending=False, inplace=True)
station_priorities = station_priorities.reset_index()
station_priorities
```

```python
import seaborn as sns
sns.set_palette("husl")
sns.barplot(y=station_priorities['station'], x=station_priorities['avg_entries'], color = "pink");
plt.xlabel("Average Weekly Entries")
plt.ylabel("MTA Station")
```
